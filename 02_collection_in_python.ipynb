--> 3. Dictionaries:

1. created using curly braces {}with  key- values pairs, or using dict()
2. it is unordered, it is supported in python 3.7+ (it maintains insertion order as an implementation details.)
3. keys must be unique and immutable but the values of key can change.
4. we can add, remove and change  key-value pairs

```python
# creating dictionay

empty_dict={}
person_dict={"name": "Asim", "age": 20, "education": "Undergraduate"}
#using dict() to create a dictionary
another_dict=dict(city="Faisalbad", university="RIPHAH")

print(empty_dict, type(empty_dict))  #type()function to print type of data.
print(person_dict)
print(another_dict)

# Accesing values by keys:
print(person_dict["education"])
print(another_dict["city"])

#get function can also be used to get values and perform operation on it. it is best to use because it return none if the key is empty or not present.

my_city_dict=another_dict.get("city")
print(my_city_dict)

print(another_dict.get("name")) # it will return none.

# to print single things of a dictionary.
print("Keys in person dictionary:", person_dict.keys())
print("Values in person dictionary:", person_dict.values())
print("Items in person dictionary:", person_dict.items())
```

Adding, removing and updating keys:

```python
person_dict={"name": "Asim Ali", "age": 20, "education": "Undergraduate"}

print(f"Actual dictionary= ", person_dict)
# we can add new key value pairs and also update the existings.

person_dict["Cast"]="Bharwana" #it will update cast key if present otherwise add it as a new key.
print(person_dict)

person_dict["age"]= 19 #it will update
print(person_dict)

#removing key pairs 

person_dict.pop("name")  # remove the age key-value pair.
print(person_dict)

#checking if a key is present or not in dictionary
if "name" in person_dict:
    print("The name key is present and its value is " , person_dict.get["name"])

# using get for it
if "age" in person_dict:
    print("The age key is present and its value is " , person_dict["age"])
```

Best practices for dictionary:

1. use it when you need fast lookup by keys.
2. keys should be descriptive and immutable. (meaning full name).
3. use get to avoid key error when a key is not present.
4. Dictionary are great for data configration, or bounding data attributes by giving a unique class or custom class.

Sets:

1. A set is collection created by using {curly braces} or set() constructor set() works with only one element.
2. we can't store duplicate data in sets.
3. Sets are mutable we can add, remove and update them.
4. Unordered
5. we can also add immutable collections in set as elements (e.g:tuple)

Creation and basic operation:

```python
#creating a set
empty_set= set()
number_set= {6,7,8,9} # Automatically remove duplicates.
mixed_set= {"John", 3, 3.34, (3, "Jeffy"), True}

print("Empty set: ", empty_set)
print("Number set: ", number_set)          # it does not keep the order of values.
print("Mixed set: ", mixed_set)
```

Adding and removing elements:

```python
print("Empty set before Add: ", empty_set)
# Adding an elenent [.add()]
empty_set.add((4,3))
print("Empty set after add: ", empty_set)

#removing element [.discard()]
empty_set.discard((4,3))
print("Empty set after removing element: ", empty_set)


# print("Empty set before Add:", empty_set)

# # Adding an element
# empty_set.add((4, 3))
# print("Empty set after add:", empty_set)

# # Removing element
# empty_set.discard((4,3))
# print("Empty set after removing element:", empty_set)
```

Set Operation:

we can perform different mathematical operations on sets (e.g: union[|], intersection [&], difference [-], symmetric_difference [^]).

```python
set_a={1, 3, 4, 6, 7}
set_b={3, 7, 9, 13}
set_c={1, 7, 3, 4, 223}
# Union: return all the element present in both sets.
print("Union: ",set_a.union(set_b).union(set_c))  #set_a | set_b |set_c

#Intersection: return the values common in all sets
print("Intersection: ", set_a & (set_b) & (set_c))

# Difference: returns the values present in first set but not in other
print("Difference: ", (set_a - set_b).difference( set_c )) # we can use both (difference or -).

#Semmetric difference: return all values that are not common
print("Symmetric Difference", (set_a ^ set_b))
print("Symmetric Difference: ", (set_a ^ set_b) ^ set_c)
```

Best practices:

1. use set to remove duplicates (e.g: test for membership one name one user).
2. only store immutable data in sets.  set is mutable but the data in it is immutable (e.g: we can add and remove values from sets but cannot change an existing value).
3. use set operations for cleaner and more efficient code when dealing with memberships and overlaps.
