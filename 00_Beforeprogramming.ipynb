04. Check is python download on your system 

        1. Open anaconda prompt from window search bar.

        2. write { cd [miniconda or file name in which anaconda is installed ]}

        3. In next step write [python -V or python --version].
        
        *if version number is displays it means python is installed and you are ready to go*

Virtual Enviroment:
virtual enviroment are different interpreter for different projects.
They help us if we have 2 or more projects but have to work on them in same time with different versions of python or same package here virtual enviroment helps us.
They provide us different workplaces to work on different projects in the same time using same package with differnt version.

1. To create a virtual enviroment.

        a. open anaconda prompt
        b. write conda 
          create -n [name of enviroment] python=version {if want to install a specific version of py.}

2. To check enviroments in the system

        {conda info -e}

3. To activate and deactivate an enviroment

        conda activate [env name]
        conda deactivare [env name]

4. To install package in an enviroment

        pip install [pkg name==version]    it tell if the version is not availiable and also give list of all availiable versions

        conda install [pkg name==version] it shows only error if version is not availiable do not provide list of availiable version

        Pip is recommented and most useful.

5. To delete a single pkg from an enviroment:

        conda remove [pkg name]    enviroment should be activated first
        conda remove  -n enviroment name [pkg name]    if enviroment is not activated in which you want to remove pkg.  {"conflicted"}

6. To delete an enviroment

        conda remove --name [env name] --all

7. To check the packages and data installed in a env

        conda list (the env should be activated first whom data you want to see)

8. in conda prompt activate the enviroment and navigate to the directory where the file are stored

        type (code .) it will  open that folder in VS code.

After creating an enviroment if you are using jupyter notebook it is important to install  ipykernel to run that enviroment error freely  select the enviroment kernel in VS code than:
