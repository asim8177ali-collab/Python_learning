# operators in python:


python provides different operators to perform calculations and computations on variables and data we have.


# 1. Assignment operators:(= and +=, -=):

the = assigns a value to a variable and +, - adds subtracts and then assigns. 

```python
# assignment operator examples:
x=343
y=34
print(x)
print(y)

y+=4
print(y)
x-=3
print(x)

y+=x
print(y)
```

# 2. Arithematic operators:

used to perform mathematical calculations:

it includes +, -, *, %, /, //, **

```python
#Arithmatic operators examples:

a, b =20, 4
print("Addition= ", a+b)
print("Subtraction= ", a-b)
print("Multiplication= ", a*b)
print("Division= ", a/b)  #Quotient
print("Modulus= ", a%b)    #Reminder
print("Floor Division= ", a//b)  #Quotient without decimals
print("Exponentiation= ", a**b)   #power of values ** means 20 ^4 = 20*20*20*20
```

# 3. Comparison operators:

used to compare values and they result only in "TRUE" and "FALSE".


```python
# Comparison operators examples:

x, y = 3, 4
print('x > y:', x>y)
print('x < y:', x<y)
print('x == y:', x==y)
print('x != y:', x!=y)
print('x >= y:', x>=y)
print('x <= y:', x<=y)
```

# 4. Logical operators:

used to combine conditions:

--> types:

AND:(returns TRUE  when both conditions are true)

OR:(returns true if atleast one condition is true)

NOT:(inverts boolean values [it returns true if value is false and vice versa.])


priority wise:  (not, and, or)

```python
# Logical operators examples:

a, b, c = True, False, False
print("a or b or c: ", a or b or c)
print("a and b and c: ", a and b and c)
print("a or b and c: ", a or b and c)
print("a and b or c: ", a and b or c)
print("Not a: ", not a)
```
