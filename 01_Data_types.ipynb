# Techniques of naming variable:
    snake_case: all small letters with underscore between them. [user_name, first_name].
    camelCase: first word small second word first letter capital all other small. [userName, firstName].
    PascalCase: first letter of each word is capital. [UserName, FirstName].

    *Through out a whole project use only one of the above technique*

# Rules for naming variables:

1. it cannot start with a number.
2. it can contain only alphabets, digits, and underscore.
3. name should be discreptive [it should show its purpose].
4. don't use reserve words [e.g; print, if, for]

    *Variable name is case sensitive*  [FIRST is not = first]
