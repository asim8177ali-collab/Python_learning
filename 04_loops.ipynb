# For loop




````python
# syntax:
    # for item in sequence:
    # process item

#example
number=[3,34,3,2,23]
for item in number:
    print(item)
````

Using enumerates() in for loops:

sometime, we need both the index and the value stored in it, so the enumerate()  function mades it easy.

````python
# example: using enumerates to view index and the value.

fruits=["apple", "pineapple", "banana", "oranges", "peach"]

for index, fruit in enumerate(fruits):
    print(f"index {index}: {fruit}")
````

````python
# example without using enumerates:

fruits=["apple", "pineapple", "banana", "oranges", "peach"]
index=0

for fruit in fruits:
    print(f"index {index}: {fruit}")
    index+=1        # here we have to make increment in index manually but enumerate() do it for us automatically.
````

# using  for loops in dictionaries:

Dictionaries are key-value pairs. we can iterate from them in three ways:

1. keys: looping over a dictionary returns its key.
2. values: Using .values()
3. key-value pairs: Using .items()



````python
# example: itrating over a dictionary

person={
    'name':'Asim',
    'city':'Faisalbad',
    'Education': 'Software Engineering',
    'status':'Single'
}

print("Iterating through Dictoinaries:")

for key in person:
    print(f"{key}: {person[key]}")
````

````python
# iterating over keys and values using item()

print("\nIterating from key-value pairs:")

for key, value in person.items():
    print(f"{key}: {value}")
````

# WHILE Loop:

the while loop runs untill a specific condition is true.

Basic syntax:

    while condition:

        executeable code block


````python
# example:

count= 0 # counter for condition

while count<10:   # the loop works until count < 10
    print("count is: ", count)
    print("I Love the Nature.")
    count+=1
````

# Loop control statements:

They allows us to alter the normal flow of the program.

1. Break: 
- terminate the program.
- useful when after meeting a specific condition no further processing or execution of program is needed.

2. Continue: 
- Skips the current iteration.
- useful to bypass certain condition.

3. Else:
- executes if loop completes normally (not terminated by *break*).
- can be used after the finishing the loop prcessing.

````python
 # using break, continue, else in a loop:

for i  in range(10): #loop from 0 to 9
    if i==3:
        print("Skipping 3.")
        continue
    
    if i==8:
        print("Breaking at 8.")
        break

    print("current number: ", i)

else:
    print("Loop executed normally.")
    
````

# Advanced Looping Techniques

Beyond the basics, Python offers several advanced techniques to make your loops more concise and efficient.

### List Comprehensions

- **What are they?**  
  A compact way to process all or part of the elements in a collection and return a list with the results.

- **Syntax Example:**

```python
[expression for item in iterable if condition]

````python
# List Comprehension to create a list of squares
squares = [x ** 2 for x in range(1, 11)]   # Compute the square for numbers 1 through 10
print("List of squares:", squares)
````

````python
# Generator Expression to create cubes
cubes = (x ** 3 for x in range(1, 11))   # Create a generator for cubes of numbers 1 through 10
print("Cubes generated:")
for cube in cubes:                       # Iterate through the generator to print each cube
    print(cube)
````

````python
# Using zip to iterate over two lists simultaneously
names = ['Alice', 'Bob', 'Charlie']   # List of names
ages = [25, 30, 35]                     # Corresponding list of ages
for name, age in zip(names, ages):      # zip pairs items from both lists together
    print(f"{name} is {age} years old")
````

````python
# Using itertools to cycle through a list
import itertools

colors = ['red', 'green', 'blue']   # List of colors
counter = 0                         # Counter to limit the loop
print("Cycling through colors using itertools.cycle:")
for color in itertools.cycle(colors):  # Cycle through colors indefinitely
    print(color)
    counter += 1
    if counter == 10:  # Stop after 6 iterations to avoid an infinite loop
        break
````
