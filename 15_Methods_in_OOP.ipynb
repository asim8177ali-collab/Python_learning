## Methods in OOP

Methods are the functions defined inside a class. They defne the behaviour of class objects.

## Types of Methods:
1. Instance Methods
2. Class Methods
3. Static Methods

## Instance Methods:
Methods that change or manage the behaviour of instances of a class. They take (self) as the first argument.



```python
# Mini challenge of instance method Circle area calculation
class Circle:
# Constructor of class to define the attributes 
    def __init__(self, radius):
        self.radius= radius
# Instance Method to calculate area of a circle
    def calculate_area(self):
        return 3.1416*self.radius**2

circle= Circle(5.3)
print(f"Area of Circle with radius {circle.radius} is {circle.calculate_area():.3f}.") # showing output to 3 decimal places
```

## Class Methods:
They work on the behaviour of class. They work on class-level state not for single object. It uses (cls) as first argument. 
To call or use a class method @classmethod is used.

Class methods are linked to the class itself, not individual instances.

Decorated with @classmethod.

Their first parameter is cls, which refers to the class.
They interact with or manipulate class-level attributes.


```python
class House:
    def __init__(self, color, age):
        self.color = color      # Instance attribute: specific color of this house
        self.age = age          # Instance attribute: how old the house is

    @classmethod
    def from_construction_year(cls, color, construction_year):
        # Class method to create a House object by calculating age from construction year
        return cls(color, 2026 - construction_year)

# Create a new house instance using the class method as a factory
my_house = House.from_construction_year("white", 2020)

# Print the age of the house instance
print(my_house.age)
```

## Static Methods = Independent Utility Tools
A static method is a method that belongs to a class but does not access or modify the class itself or any instance of the class. Unlike instance methods or class methods, static methods do not take self (the instance) or cls (the class) as their first parameter.

Static methods behave like regular functions, but they are logically grouped inside a class because they perform operations relevant to the class’s purpose.

They are typically used for utility tasks such as validations, calculations, or helper functions that don’t rely on the state of the class or its instances.

Static methods are defined using the @staticmethod decorator.

These are utility functions related to class.

## Using the static method to validate area before creating a House instance


```python
class House:

    def __init__(self, color, area):
        self.color = color
        self.area = area  # Area in square meters

    @staticmethod
    def is_valid_area(area):
        # Static method to validate if the area value is reasonable (positive)
        return area > 0
    
print(House.is_valid_area(120))   # Expected output: True
print(House.is_valid_area(-50))   # Expected output: False

 
```

```python
class MathTool:
    pi = 3.14159
    def __init__(self, radius):
        self.radius = radius

    # Instance method to calculate the area of a circle
    def area(self):
        return MathTool.pi * self.radius * self.radius
    
    # Class method to create a MathTool object from diameter
    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)
    
    # Static method to validate if a radius is positive
    @staticmethod
    def is_valid_radius(radius):
        return radius > 0

# Example usage:
tool = MathTool(5)
print(tool.area())  # Output: 78.53975
tool_from_diameter = MathTool.from_diameter(10)
print(tool_from_diameter.radius)  # Output: 5.0
print(MathTool.is_valid_radius(1))  # Output: False
```

# Methods in Python classes

```python
# Instance Methods:
# Rectangle class to return and print perimeter
class Rectangle:

    def __init__(self, height, width):
        self.height = height
        self.width = width

    def perimeter(self):
        peri= 2*(self.height + self.width)
        return peri

rectangle_perimeter=Rectangle(3 , 2) 
print(f"Rectangle Perimeter= {rectangle_perimeter.perimeter()}")

print("------------------------------------------------")

# Class Methods
# Class for date
class Date:

    def __init__(self, year, month, day):
        self.year= year 
        self.month= month 
        self.day= day 

    @classmethod
    def from_string(cls,date_string):
        year, month, day= date_string.split("-")
        return cls(int(year), (month), int(day))
    
date=Date.from_string("2026-June-7")
print(f"Year= {date.year} | Month= {date.month} | Day={date.day}")

print("------------------------------------------------")

# Static Methods
# Palindrome checker as a string Helper
class StringHelper:

    def __init__(self, string):
        self.string = string

    @staticmethod
    def palindrome_checker(string):
        cleaned = string.lower()              # 👈 handle capitals
        if cleaned == cleaned[::-1]:
            print(f"✅ '{string}' is a Palindrome!")
        else:
            print(f"❌ '{string}' is not a Palindrome!")

pali = StringHelper("Python")
pali.palindrome_checker("Racecar")   # ✅ Palindrome
pali.palindrome_checker("Python")    # ❌ Not Palindrome
pali.palindrome_checker("Madam")     # ✅ Palindrome

print("------------------------------------------------")

# Combined Example
# Class toolbox with instance method (add_tool), class method (get_total_toolbox) and static method (is_valid_tool)

class ToolBox:
    toolbox_count=0
    def __init__ (self, tools=None):
        self.tools= tools if tools else []
        ToolBox.toolbox_count+=1

    def add_tool(self, tool):
        if ToolBox.is_valid_tool(tool, self.tools):
            print(f"❌ {tool} already exists.")
        else:
            self.tools.append(tool)
            print(f"{tool}✅ Tool added")
    
    def show_tools(self):
        print(f"Tools: {self.tools}")

    @classmethod
    def get_total_toolboxes(cls):
        print(f"There are {cls.toolbox_count} tool box created.")
    
    @staticmethod 
    def is_valid_tool(tool_name, tools):
        return tool_name in tools          

# Testing
box1 = ToolBox(['Nails'])
box2 = ToolBox(['Screwdriver', 'Wrench'])

box1.add_tool('Tester')        # new tool
box1.add_tool('Hammer')        # already exists
box1.show_tools()

ToolBox.get_total_toolboxes()

print(ToolBox.is_valid_tool('Hammer', box1.tools))
print(ToolBox.is_valid_tool('Wrench', box1.tools))
```
